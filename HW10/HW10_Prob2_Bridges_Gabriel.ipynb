```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
```

```python
t, sig, dw = np.loadtxt("Data3.csv",skiprows=4 , unpack=True, delimiter=",")
```

```python
def oscDecay(t, A, B, C, tau, omega):
    d = [A*(1+B*np.cos(omega*t[0]))*np.exp(-(t[0]**2)/(2*(tau**2))) + C]
    for i in range(1,len(t)):
        d += [A*(1+B*np.cos(omega*t[i]))*np.exp(-(t[i]**2)/(2*(tau**2))) + C]
    return d
def chiSqr(t, sig, dw,po):
    chiSquared = 0
    opt, pcov = scipy.optimize.curve_fit(oscDecay, t, sig, p0=po)
    for n in range(0, len(sig)):
        chiSquared += ((sig[n]-oscDecay(t,*opt)[n])/dw[n])**2
    return chiSquared
```

```python
plt.errorbar(t,sig,yerr=dw,fmt = 'ro', ecolor = 'black',capsize=2,elinewidth=1.3,markeredgecolor='black',markersize=5)
plt.plot(t, oscDecay(t,12,1,20,15,0.75))
plt.ylim(15, 45)
plt.xlim(0,40)
plt.show()
```

```python
po=[12,1,20,15,0.75]
opt, pcov = scipy.optimize.curve_fit(oscDecay, t, sig, p0=po)
```

```python
plt.errorbar(t,sig,yerr=dw,fmt = 'ro', ecolor = 'black',capsize=2,elinewidth=1.3,markeredgecolor='black',markersize=5)
plt.plot(t, oscDecay(t,*opt))
plt.text(20, 40, "$\chi^2={0:.3f}$".format(chiSqr(t, sig, dw, po)))
plt.text(20, 37, "$A={0:.3f}$".format(opt[0]))
plt.text(20, 34, "$B={0:.3f}$".format(opt[1]))
plt.text(33, 40, "$C={0:.3f}$".format(opt[2]))
plt.text(33, 37, "$\\tau={0:.3f}$".format(opt[3]))
plt.text(33, 34, "$\omega={0:.3f}$".format(opt[4]))
plt.ylim(15, 45)
plt.xlim(0,40)
plt.show()
```

```python
def chiSqrPlot(t, sig, dw,po):
    po[4] = 0.05
    chiPlt = [chiSqr(t, sig, dw, po)]
    while (po[4] < 3.90):
        po[4] += 0.05
        chiPlt += [chiSqr(t, sig, dw, po)]
    return chiPlt
```

```python
omega = np.linspace(0.05, 3.95, 79)
plt.plot(omega,chiSqrPlot(t, sig, dw, po))
plt.show()
```
